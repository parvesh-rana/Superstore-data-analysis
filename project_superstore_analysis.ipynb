```python
state_profit = ds.groupby('State')['Profit'].sum().reset_index()
```

```python
state_profit_sorted = state_profit.sort_values("Profit", ascending=False)
```

```python
top_20_states = state_profit_sorted.head(20)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12, 6))
plt.bar(top_20_states['State'], top_20_states['Profit'])
plt.xlabel('State')
plt.ylabel('Profit')
plt.title('Top 20 Profit States in Superstore Dataset')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
